```python
df.groupby(['D'])
```

```python
df.groupby(['D']).Duration.agg([('Min' , 'min'), ('Median', 'median'), ('Max', 'max')])
```

```python
def percentile(n):
    def percentile_(x):
        return np.percentile(x, n)
    percentile_.__name__ = 'percentile_%s' % n
    return percentile_
```

```python
df.groupby(['D']).Duration.agg([('Min' , 'min'), ('Q1', percentile(25)), ('Median', 'median'), ('Q3', percentile(75)), ('Max', 'max')])
```
